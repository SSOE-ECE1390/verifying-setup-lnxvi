Imports

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Load and View image

```python
# load image
image_path = "Data/Apollo_11_Launch.jpg"
image = cv2.imread(image_path, 0)  # grayscale

if image is None:
    print("Image does not exist. Check path")
else:
    print("Image found successfully")
    plt.imshow(image, cmap='gray')  # view original image
```

Crop Image

```python
cropped_image = image[25:650, 425:775]
plt.imshow(cropped_image, cmap='gray')
```

Edit Brightness

```python
matrix = np.ones(cropped_image.shape, dtype="uint8") * 75
brighter_image = cv2.add(cropped_image, matrix)
plt.imshow(brighter_image, cmap='gray')
```

Rotate

```python
rotated_image = cv2.rotate(brighter_image, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(rotated_image, cmap='gray')
```

Save

```python
save_path = "Data/Apollo_11_Launch_MODIFIED.jpg"
cv2.imwrite(save_path, rotated_image)
```
